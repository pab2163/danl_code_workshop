## Selecting Columns

```python
# select columns by defining a list of columns
df_clean = df[["subjNum", "groupStatus", "adhd", "hitRate", "faRate", "dprime"]]
```

## Filtering rows!

```python
df_clean.head()
```

```python
# test whether each value in the subjNum column is equal to 10
df_clean.subjNum == 10
```

```python
# filter to get only the rows of df_clean, where subjNum is 10
df_clean[df_clean.subjNum == 10]
```

```python
# filter to get only the rows of df_clean, where subjNum is NOT 10
df_clean[df_clean.subjNum != 10]
```

```python
# what values are in the groupStatus column?
df_clean.groupStatus.unique()
```

```python
df_clean[df_clean.groupStatus == 'LMM']
```

```python
df_no_outliers = df_clean[df_clean.hitRate >= .25]
```

```python
df_no_outliers.describe()
df_clean.describe()
```

```python
# multiple filtering conditions
df_no_outliers = df_clean[(df_clean.hitRate >= .25) | (df_clean.groupStatus == 'LMM')]
```

```python
df_no_outliers.describe()
```

```python
df_no_outliers.groupStatus.unique()
```

```python
df_clean.head()
```

## remove subjNum of 10, 5, and 2


```python
subs_to_remove = [10, 5, 2]


# give ONLY those participants using the .isin() method
df_clean[df_clean.subjNum.isin(subs_to_remove)]
```

```python
# remove those participants
df_clean[-df_clean.subjNum.isin(subs_to_remove)]
```

# Filter for hitrate between .25 and .75

```python
df_clean[(df_clean.hitRate >= .25) & (df_clean.hitRate <= .75)]
```

```python
df_clean[(df_clean.hitRate <= .25) | (df_clean.hitRate >= .75)]
```
