# Picking certain columns

2 main ways to reference dataframe columns in pandas
1. `music['danceability']`
2. `music.danceability`

This is like in R when we would do something like `music$danceability`

```python
# gives you back one column JUST as a column (pandas series)
print(type(music['danceability']))
music['danceability']
```

```python
# gives you back a DATAFRAME that has one columns
print(type(music[['danceability']]))
music[['danceability']]
```

```python
music.danceability
```

```python
# if we ask for a LIST of columns, we get back a dataframe with those columns
music[['danceability', 'key']]
```

```python
column_list_1 = ['key', 'rank', 'artist']
column_list_1
```

```python
music[column_list_1]
```

```python
# specificy columns to get rid of
music.drop(columns = column_list_1)
```
