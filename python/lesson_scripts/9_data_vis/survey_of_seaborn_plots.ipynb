# Import seaborn

```python
import pandas as pd

# seaborn shorthand: sns
import seaborn as sns

# makes sure plots show up inline in the jupyter notebook
%matplotlib inline
```

# Import some data

```python
df = pd.read_csv('../../../data/familiar_music_database.csv')
```

```python
df.info()
```

```python
df.head()
```

# Histograms

Looking at distributions of single continuous variables

```python
sns.distplot(df.tempo)
```

```python
# turn off the smooth density distribution
sns.distplot(df.valence, kde=False)
```

# Scatter plots

great for exploring relationships between continuous variables!

```python
sns.scatterplot(data = df, x = 'tempo', y = 'valence')
```

```python
sns.scatterplot(data = df, x = 'tempo', y = 'valence', hue = 'year')
```

```python
sns.scatterplot(data = df, x = 'tempo', y = 'valence', hue = 'genre')
```

# Jointplots: scatterplot + histograms

```python
sns.jointplot(data = df, x = 'year', y = 'tempo')
```

# Categorical Plots

```python
sns.catplot(data = df, x = 'genre', y = 'tempo', s = 2, alpha = .3)
```

```python
sns.catplot(data = df, x = 'genre', y = 'tempo', kind = 'box')
```

```python
sns.boxplot(data = df, x = 'genre', y = 'tempo')
```

```python
sns.catplot(data = df, x = 'genre', y = 'tempo', kind = 'violin')
```

```python
sns.catplot(data = df, x = 'genre', y = 'tempo', kind = 'point', join = False)
```

```python
sns.countplot(df.genre)
```

# Plot with linear model fit to the data

sns.lmplot()

```python
sns.lmplot(data = df, x = 'year', y = 'tempo', hue = 'genre')
```

# Splitting into panels

```python
sns.lmplot(data = df, x = 'year', y = 'tempo', hue = 'genre', col = 'genre')
```

```python
sns.lmplot(data = df, x = 'year', y = 'tempo', hue = 'genre', col = 'genre', row = 'mode')
```
